````python
import difflib

file1_path = "loose_1.txt"
file2_path = "loose_2.txt"

# Read playlists
with open(file1_path, 'r', encoding='utf-8') as f1:
    file1_lines = [line.strip() for line in f1 if line.strip()]  # remove blank lines
with open(file2_path, 'r', encoding='utf-8') as f2:
    file2_lines = [line.strip() for line in f2 if line.strip()]

print(file1_lines)
# Create diff
diff = difflib.unified_diff(
    file1_lines,
    file2_lines,
    fromfile='Old Playlist',
    tofile='New Playlist',
    lineterm=''
)

# Format for Discord code block
discord_message = "```diff\n"
for line in diff:
    if line.startswith('+') and not line.startswith('+++'):
        discord_message += f"{line}\n"   # green for additions
    elif line.startswith('-') and not line.startswith('---'):
        discord_message += f"{line}\n"   # red for removals
discord_message += "```"

print(discord_message)
````

````python

# Read playlists (remove empty lines and trim)
with open(file1_path, 'r', encoding='utf-8') as f1:
    old_playlist = [line.strip() for line in f1 if line.strip()]
with open(file2_path, 'r', encoding='utf-8') as f2:
    new_playlist = [line.strip() for line in f2 if line.strip()]

# Build message
discord_message = "```diff\n"

# Check for removed or changed positions
for idx, song in enumerate(old_playlist, start=1):
    if song not in new_playlist:
        discord_message += f"- {idx:02d}. {song}\n"

# Check for added songs
for idx, song in enumerate(new_playlist, start=1):
    if song not in old_playlist:
        discord_message += f"+ {idx:02d}. {song}\n"

discord_message += "```"

print(discord_message)
````
