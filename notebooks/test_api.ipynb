```python
from openai import OpenAI
from dotenv import load_dotenv
import os

api_key = os.getenv("NVIDIA_API_KEY")

system_message = "On va te donner un message d'une personne extraite d'une conversation discord. Ton rôle est d'y répondre de manière très sarcastique et agressive."
# system_message = "You are a AI assistant. You will be given a message from someone in a chat group. Your role to answer it in a very sarcastic way, using the language of the message."
user_name = "Jean-Paul Théodule"
user_message = "En vrai je pense juste voc faut vraiment que j'avance sur ma fig"

messages = [
    {"role":"user", "content":f"{system_message}\n\nVoici le message de {user_name} : {user_message}"},
]


client = OpenAI(
  base_url = "https://integrate.api.nvidia.com/v1",
  api_key = api_key
)


completion = client.chat.completions.create(
  model="meta/llama-3.1-70b-instruct",
  messages=messages,
  temperature=0.2,
  top_p=0.7,
  max_tokens=1024,
  # stream=True
)

print(completion.choices[0].message.content)

# for chunk in completion:
#   if chunk.choices[0].delta.content is not None:
#     print(chunk.choices[0].delta.content, end="")

```

```python
# load_dotenv("../.env")
from datetime import datetime, timedelta
expiration_date = os.getenv("NVIDA_EXPIRATION_DATE")

# Convert it to a datetime object
input_date = datetime.strptime(expiration_date, "%d/%m/%Y")

# Get the current date (no time)
now = datetime.now() - timedelta(days=1)
print(now)

# Compare the dates
if input_date > now:
    print("The date is in the future.")
elif input_date < now:
    print("The date is in the past.")
else:
    print("The date is today.")
```
